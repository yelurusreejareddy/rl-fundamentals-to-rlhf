# Breadth-First Search:

```python
def breadth_first_search(start_state, target_state):
    frontier_stack = Stack()
    visited_states = {}
    frontier_queue = Queue()
    root_node = Node(State(start_state), dep=0, p_cost=0)
    frontier_queue.push(root_node)
    step_counter = 0
    max_space = 0

    while frontier_queue.queue and step_counter < 21:
        max_space = max(max_space, frontier_queue.space())
        current_node = frontier_queue.pop()
        frontier_stack.push(current_node)
        
        print(f"count={step_counter}, depth={current_node.status['depth']}, cost={current_node.status['path_cost']}")
        for i in range(0, len(current_node.state.st_list), 3):
            print(current_node.state.st_list[i], current_node.state.st_list[i+1], current_node.state.st_list[i+2])

        if current_node.state.st_list == target_state:
            print(current_node.state.st_list)
            break  
        else:
            if tuple(current_node.state.st_list) not in visited_states:
                visited_states[tuple(current_node.state.st_list)] = True
                successors = current_node.find_successors()
                for child_node in successors:
                    if tuple(child_node.state.st_list) not in visited_states:
                        frontier_queue.push(child_node)

        step_counter += 1

    print(f"Max Space Used: {max_space}")
    """
    print("Tracing Back Path")

    while frontier_stack.stack:
        step_counter -= 1
        traceback_node = frontier_stack.pop()
        print(f"count={step_counter}, depth={traceback_node.status['depth']}, cost={traceback_node.status['path_cost']}")
        for i in range(0, len(traceback_node.state.st_list), 3):
            print(traceback_node.state.st_list[i], traceback_node.state.st_list[i+1], traceback_node.state.st_list[i+2])
"""
breadth_first_search(initial, goal)
```

# Depth-First Search:

```python
def depth_first_search(start_state, target_state):
    frontier_stack = Stack()
    visited_states = {}
    root_node = Node(State(start_state), dep=0, p_cost=0)
    frontier_stack.push(root_node)
    step_counter = 0
    max_space = 0

    while frontier_stack and step_counter < 21:
        max_space = max(max_space, frontier_stack.space())
        current_node = frontier_stack.pop()
        
        print(f"count={step_counter}, depth={current_node.status['depth']}, cost={current_node.status['path_cost']}")
        for i in range(0, len(current_node.state.st_list), 3):
            print(current_node.state.st_list[i], current_node.state.st_list[i+1], current_node.state.st_list[i+2])

        if current_node.state.st_list == target_state:
            break  
        
        if tuple(current_node.state.st_list) not in visited_states:
            visited_states[tuple(current_node.state.st_list)] = True
            successors = current_node.find_successors()
            for successor in successors:
                if tuple(successor.state.st_list) not in visited_states:
                    frontier_stack.push(successor)

        step_counter += 1

    print(f"Max Space Used: {max_space}")

depth_first_search(initial, goal)
```

# Uniform Cost Search:

```python
def uniform_cost_search(start_state, target_state):
    frontier_queue = PriorityQueue()  
    visited_states = {}
    root_node = Node(State(start_state), act="ucs", dep=0, p_cost=0)
    frontier_queue.push(root_node, root_node.status["path_cost"])
    step_counter = 0
    max_space = 0
    
    while frontier_queue and step_counter < 21:
        current_node = frontier_queue.pop()
        max_space = max(max_space, frontier_queue.space())

        print(f"count={step_counter}, depth={current_node.status['depth']}, cost={current_node.status['path_cost']}")

        for i in range(0, len(current_node.state.st_list), 3):
            print(current_node.state.st_list[i], current_node.state.st_list[i + 1], current_node.state.st_list[i + 2])
        
        if current_node.state.st_list == target_state:
            print("Goal Achieved:", current_node.state.st_list)
            break  
        
        if tuple(current_node.state.st_list) not in visited_states:
            visited_states[tuple(current_node.state.st_list)] = True
            successors = current_node.find_successors()
            for successor in successors:
                if tuple(successor.state.st_list) not in visited_states:
                    frontier_queue.push(successor, successor.status["path_cost"])
        
        step_counter += 1
    
    print(f"Max Space Used: {max_space}")

uniform_cost_search(initial, goal)
```

# Greedy Best-First Search:

```python
def greedy_best_first_search(start_state, target_state):
    frontier_queue = PriorityQueue()  
    visited_states = {}
    root_node = Node(State(start_state), act="greedy", dep=0, p_cost=0)
    frontier_queue.push(root_node, root_node.state.misplaced_count(root_node.state.st_list))
    step_counter = 0
    max_space = 0
    
    while frontier_queue and step_counter < 21:
        current_node = frontier_queue.pop()
        max_space = max(max_space, frontier_queue.space())

        print(f"count={step_counter}, depth={current_node.status['depth']}, cost={current_node.status['path_cost']}")

        for i in range(0, len(current_node.state.st_list), 3):
            print(current_node.state.st_list[i], current_node.state.st_list[i + 1], current_node.state.st_list[i + 2])
        
        if current_node.state.st_list == target_state:
            print("Goal Achieved:", current_node.state.st_list)
            break  
        
        if tuple(current_node.state.st_list) not in visited_states:
            visited_states[tuple(current_node.state.st_list)] = True
            successors = current_node.find_successors()
            for successor in successors:
                if tuple(successor.state.st_list) not in visited_states:
                    frontier_queue.push(successor, successor.state.misplaced_count(successor.state.st_list))
        
        step_counter += 1

    print(f"Max Space Used: {max_space}")

greedy_best_first_search(initial, goal)
```

# A* Search:

```python
def a_star_search(start_state, target_state):
    priority_queue = PriorityQueue()  
    visited_states = {}
    root_node = Node(State(start_state), act="a_star", dep=0, p_cost=0)
    priority_queue.push(root_node, root_node.state.misplaced_count(root_node.state.st_list) + root_node.status["path_cost"])
    
    step_counter = 0
    max_space = 0
    
    while priority_queue and step_counter < 21:
        current_node = priority_queue.pop()
        max_space = max(max_space, priority_queue.space())

        print(f"count={step_counter}, depth={current_node.status['depth']}, cost={current_node.status['path_cost']}")

        for i in range(0, len(current_node.state.st_list), 3):
            print(current_node.state.st_list[i], current_node.state.st_list[i + 1], current_node.state.st_list[i + 2])
        
        if current_node.state.st_list == target_state:
            print("Goal Achieved:", current_node.state.st_list)
            break  
        
        if tuple(current_node.state.st_list) not in visited_states:
            visited_states[tuple(current_node.state.st_list)] = True
            successors = current_node.find_successors()
            for successor in successors:
                if tuple(successor.state.st_list) not in visited_states:
                    priority_queue.push(successor, successor.state.misplaced_count(successor.state.st_list) + successor.status["path_cost"])
        
        step_counter += 1

    print(f"Max Space Used: {max_space}")

a_star_search(initial, goal)
```

# A* Search (Manhattan Distance):

```python
def a_star_manhattan(initial, goal):
    
    def heuristic_function(node):
        total_distance = 0
        for index, tile in enumerate(node.state.st_list):
            if tile != 0:  
                goal_index = goal.index(tile)
                current_pos = (index // 3, index % 3)
                goal_pos = (goal_index // 3, goal_index % 3)
                total_distance += manhattanDistance(current_pos, goal_pos)
        
        return total_distance + node.status["path_cost"]

    frontier = PriorityQueueWithFunction(heuristic_function)
    explored = set()
    
    root_node = Node(state=State(initial), act=None, dep=0, p_cost=0)
    frontier.push(root_node)
    
    step_count = 0  
    max_space = 0  

    while not frontier.isEmpty() and step_count < 21:
        current_node = frontier.pop()
        max_space = max(max_space, len(frontier.heap))

        print(f"count={step_count}, depth={current_node.status['depth']}, cost={current_node.status['path_cost']}")
        
        for i in range(0, len(current_node.state.st_list), 3):
            print(current_node.state.st_list[i], current_node.state.st_list[i + 1], current_node.state.st_list[i + 2])

        if current_node.state.st_list == goal:
            print("Goal Found:", current_node.state.st_list)
            break  
        
        if tuple(current_node.state.st_list) not in explored:
            explored.add(tuple(current_node.state.st_list))
            for child_node in current_node.find_successors():
                if tuple(child_node.state.st_list) not in explored:
                    frontier.push(child_node)
        
        step_count += 1

    print(f"Max Space Used: {max_space}")

a_star_manhattan(initial, goal)
```
